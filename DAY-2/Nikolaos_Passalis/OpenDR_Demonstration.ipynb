Pose Estimation Demonstration

First, we need to load our model. In this tutorial we are using the OpenPose estimator from OpenDR:


```python
import sys
sys.path.append("/content/openDr/lib/python3.8/site-packages")
```

```python
from opendr.perception.pose_estimation import LightweightOpenPoseLearner
```

We need to create our estimator:


```python
 pose_estimator = LightweightOpenPoseLearner(device='cuda', num_refinement_stages=2)
```

Note that we can alter the device (e.g., 'cpu', 'cuda', etc.), on which the model runs, the number of refinement stages that we are using, as well as a number of additional parameters that can make our model either faster or more accurate.

After creating our model, we need to download and load the pre-trained weights:

```python
pose_estimator.download(path=".", verbose=True)
pose_estimator.load("openpose_default")
```

We will also download one sample image for testing:

```python
pose_estimator.download(path=".", mode="test_data")
```

Let's read this image:

```python
from os.path import join
import cv2
from opendr.engine.data import Image

image_path = join("temp", "dataset", "image", "000000000785.jpg")
img = Image.open(image_path)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.imshow(img.opencv()[:, :, [2, 0, 1]])
```

We are now ready to use our model! The only thing that we have to do is to pass the image through the model. Note that there are standard data types supported by OpenDR. However, OpenDR also understands common data types (e.g,. OpenCV images) and automatically converts them into the most appropriate format:

```python
poses = pose_estimator.infer(img)
```

```python
print(poses)
```

What can we do with these poses?

```python
print(poses[0].data)
```

```python
from opendr.engine.target import Pose
print(Pose.kpt_names)
```

```python
print(poses[0])
```

```python
print(poses[0]['nose'])
```

As we can see, the inference algorithm ran correctly and returned a list of FilteredPose objects (one for each detected pose). We can easily visualize the results:

```python
from opendr.perception.pose_estimation.lightweight_open_pose.utilities import draw
img = img.opencv()
for pose in poses:
        draw(img, pose)
plt.imshow(img[:,:, [2, 1, 0]])
```

# Using OpenDR for a common Face Recognition pipeline (Face Detection + alignment + Face Recognition)

## Initialize the learner classes, one for Face Detection and one for Face Recognition. Download the available pretrained models, load them and create the database of known persons.

```python
import sys

from opendr.perception.face_recognition import FaceRecognitionLearner
from opendr.perception.object_detection_2d import RetinaFaceLearner
from opendr.perception.object_detection_2d.datasets.transforms import BoundingBoxListToNumpyArray
from opendr.perception.object_detection_2d import draw_bounding_boxes

from os.path import join
from opendr.engine.data import Image
from google.colab.patches import cv2_imshow
import cv2
import os  
```

```python
download_path ='.'
facedetector = RetinaFaceLearner(backbone='mnet', device='cuda')
facedetector.download(download_path, mode="pretrained")
facedetector.load(os.path.join(download_path, "retinaface_mnet"))
```

```python
recognizer = FaceRecognitionLearner(device='cpu', backbone='ir_50', mode='backbone_only')
recognizer.download(path=download_path)
recognizer.load(download_path)
dataset_path = '/content/OpenDR_Workshop/aligned'
save_path='/content/database'
recognizer.fit_reference(path=dataset_path, save_path=save_path, create_new=True)
```

## Show an original image

```python
image_path ='/content/OpenDR_Workshop/infer/all/1.jpg'
img = Image.open(image_path)
%matplotlib inline
from matplotlib import pyplot as plt
plt.imshow(img.opencv()[:, :, [2, 1, 0]])
```

## Detect any faces in the image and draw the corresponding bounding boxes, using OpenDR's RetinaFace module.

```python
import cv2
from google.colab.patches import cv2_imshow
font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 0.5
thickness = 2

frame = Image.open("/content/OpenDR_Workshop/infer/all/1.jpg")
frame = Image.opencv(frame)
bounding_boxes = facedetector.infer(frame)
frame = draw_bounding_boxes(frame, bounding_boxes, facedetector.classes, show=False)

cv2_imshow(frame)
```

```python
print(bounding_boxes)
```

## Detect any faces in the image, draw the corresponding bounding boxes and use the cropped sections of the image with a face to recognize the person and draw the resulting name.


```python
import cv2
from google.colab.patches import cv2_imshow
font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 0.5
thickness = 2

frame = Image.open("/content/OpenDR_Workshop/infer/all/1.jpg")
frame = Image.opencv(frame)
bounding_boxes = facedetector.infer(frame)
if bounding_boxes:
    bounding_boxes_ = BoundingBoxListToNumpyArray()(bounding_boxes)
    boxes = bounding_boxes_[:, :4]
    for idx, box in enumerate(boxes):
        (startX, startY, endX, endY) = int(box[0]), int(box[1]), int(box[2]), int(box[3])
        img = frame[startY:endY, startX:endX]
        result = recognizer.infer(img)
        if result.description != 'Not found':
            color = (0, 255, 0)
        else:
            color = (0, 0, 255)
        frame = cv2.rectangle(frame, (startX, startY), (endX, endY), color, thickness)
        frame = cv2.putText(frame, result.description, (startX, endY + 20), font,
                          fontScale, color, thickness, cv2.LINE_AA)

cv2_imshow(frame)       
```

## Use the same example as above, but now we initialize a learner class for Pose Estimation as well, download and load the available pretrained model and detect the poses in the given image as well.
